```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'Product': ['A', 'B', 'A', 'B', 'C', 'C', 'B', 'A'],
    'Sales': [150, 200, 300, 400, 250, 180, 220, 310],
    'Quantity': [10, 15, 20, 25, 12, 14, 16, 18],
}
```

```python
df=pd.DataFrame(data)
print("Sample dataset :\n",df)
```

```python
# Step 2: Grouping and Aggregation
# Aggregating Sales by Region (Sum Aggregation)
sales_by_region_sweety = df.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region:\n", sales_by_region_sweety)
```

```python
# Aggregating Sales and Quantity by Product (Mean Aggregation)
mean_by_product_sweety = df.groupby('Product')[['Sales', 'Quantity']].mean()
print("\nMean Sales and Quantity by Product:\n", mean_by_product_sweety)
```

```python
# Aggregating Count of Sales by Region (Count Aggregation)
count_by_region_sweety = df.groupby('Region')['Sales'].count()
print("\nCount of Sales Records by Region:\n", count_by_region_sweety)
```

```python
# Step 3: Multi-Level Aggregation
# Aggregating Sales by Region and Product
multi_level_agg_sweety = df.groupby(['Region', 'Product'])['Sales'].sum()
print("\nSales by Region and Product:\n", multi_level_agg_sweety)

     
```

#### Objective:
#### To understand and implement:
#### Time Aggregation: Aggregating data over different time periods(eg:monthly,yearly)
#### Spatial Aggregation : Aggregating  data by spatial attribute(eg:by region,city)

```python
import pandas as pd
```

```python
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West'],
    'City': ['City1', 'City2', 'City3', 'City4', 'City1', 'City2', 'City3', 'City4'],
    'Product': ['A', 'B', 'A', 'B', 'C', 'C', 'B', 'A'],
    'Sales': [150, 200, 300, 400, 250, 180, 220, 310],
    'Quantity': [10, 15, 20, 25, 12, 14, 16, 18],
    'Date': pd.to_datetime(['2024-01-01', '2024-01-02', '2024-02-01', '2024-02-03',
                            '2024-03-01', '2024-03-02', '2024-04-01', '2024-04-03'])
}
```

```python
df_sweety = pd.DataFrame(data)
print("Extended Dataset:\n", df_sweety)

     
```

```python
# Time Aggregation
# Step2: Set Date Column as Index
df_sweety.set_index('Date', inplace=True)
```

```python
# Aggregating Sales by Month
monthly_sales_sweety = df_sweety.resample('M')['Sales'].sum()
print("\nTotal Sales by Month:\n", monthly_sales_sweety)
```

```python
# Aggregating Sales by Quarter
quarterly_sales_sweety = df_sweety.resample('Q')['Sales'].sum()
print("\nTotal Sales by Quarter:\n", quarterly_sales_sweety)
```

```python
# Aggregating Sales by Year
yearly_sales_sweety = df_sweety.resample('Y')['Sales'].sum()
print("\nTotal Sales by Year:\n", yearly_sales_sweety)
```

```python
# Reset Index to Restore Original Structure
df_sweety.reset_index(inplace=True)
```

```python
#Spatial Aggregation
#Step3:Aggregating Sales by Region
sales_by_region_sweety=df_sweety.groupby('Region')['Sales'].sum()
print("\nTotal Sales by Region:\n", sales_by_region_sweety)
```

```python
# Aggregating Sales by City
sales_by_city_sweety = df_sweety.groupby('City')['Sales'].sum()
print("\nTotal Sales by City:\n", sales_by_city_sweety)
```

```python
# Aggregating Sales by Region and City
sales_by_region_city_sweety = df_sweety.groupby(['Region', 'City'])['Sales'].sum()
print("\nTotal Sales by Region and City:\n", sales_by_region_city_sweety)
```

```python
#Step 4: Export Spatial Aggregation Results
sales_by_region_city_reset_sweety=sales_by_region_city_sweety.reset_index()
sales_by_region_city_reset_sweety.to_csv("spatial_aggregation.csv",index= False)
print("\nSpatial aggregation data saved to 'spatial_aggregation'")
```
